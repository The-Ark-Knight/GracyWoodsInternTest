```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", 
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", 
           "hours-per-week", "native-country", "income"]
df = pd.read_csv(url, names=columns, na_values=" ?", skipinitialspace=True)
```

```python
#checking how many rows have null values

count = 0
for index, row in df.iterrows():
    if (row["workclass"] == "?" or row["occupation"] == "?") or row["native-country"] == "?":
        count += 1
        
count
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plotting graphs for further understanding

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of Income
plt.figure(figsize=(10, 6))
sns.countplot(x='income', data=df)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()

# Box plot of Age vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='income', y='age', data=df)
plt.title('Box Plot of Age vs Income')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()

# Count plot of Education level
plt.figure(figsize=(14, 8))
sns.countplot(y='education', data=df, order=df['education'].value_counts().index)
plt.title('Education Level Distribution')
plt.xlabel('Count')
plt.ylabel('Education')
plt.show()

# Hours per week by income
plt.figure(figsize=(10, 6))
sns.histplot(df, x="hours-per-week", hue="income", multiple="stack", bins=30)
plt.title('Hours per Week by Income')
plt.xlabel('Hours per Week')
plt.ylabel('Count')
plt.show()

# Capital Gain by Income
plt.figure(figsize=(10, 6))
sns.histplot(df, x="capital-gain", hue="income", multiple="stack", bins=30)
plt.title('Capital Gain by Income')
plt.xlabel('Capital Gain')
plt.ylabel('Count')
plt.show()

# Occupation count plot
plt.figure(figsize=(14, 8))
sns.countplot(y='occupation', data=df, order=df['occupation'].value_counts().index)
plt.title('Occupation Distribution')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.show()

# Race count plot
plt.figure(figsize=(10, 6))
sns.countplot(y='race', data=df, order=df['race'].value_counts().index)
plt.title('Race Distribution')
plt.xlabel('Count')
plt.ylabel('Race')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='hours-per-week', hue='income', data=df)
plt.title('Age vs. Hours per Week by Income')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.legend(title='Income')
plt.show()

# Education Level vs. Hours per Week by Income
plt.figure(figsize=(14, 8))
sns.boxplot(x='hours-per-week', y='education', hue='income', data=df)
plt.title('Education Level vs. Hours per Week by Income')
plt.xlabel('Hours per Week')
plt.ylabel('Education')
plt.legend(title='Income')
plt.show()

# Workclass vs. Hours per Week by Income
plt.figure(figsize=(14, 8))
sns.boxplot(x='hours-per-week', y='workclass', hue='income', data=df)
plt.title('Workclass vs. Hours per Week by Income')
plt.xlabel('Hours per Week')
plt.ylabel('Workclass')
plt.legend(title='Income')
plt.show()

# Age vs. Capital Gain by Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='capital-gain', hue='income', data=df)
plt.title('Age vs. Capital Gain by Income')
plt.xlabel('Age')
plt.ylabel('Capital Gain')
plt.legend(title='Income')
plt.show()

# Occupation vs. Income
plt.figure(figsize=(14, 8))
sns.countplot(y='occupation', hue='income', data=df, order=df['occupation'].value_counts().index)
plt.title('Occupation vs. Income')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.legend(title='Income')
plt.show()

# Race vs. Education Level by Income
plt.figure(figsize=(14, 8))
sns.countplot(y='race', hue='education', data=df, order=df['race'].value_counts().index)
plt.title('Race vs. Education Level by Income')
plt.xlabel('Count')
plt.ylabel('Race')
plt.legend(title='Education Level')
plt.show()
```

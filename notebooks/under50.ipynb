```python
from itertools import combinations
from sklearn.preprocessing import PolynomialFeatures

# Define primary features
primary_features = ['age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss']
workclass_feature = ['workclass_Private']

# Generate polynomial features for interactions up to degree 2 (pairwise interactions)
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
interaction_features_2 = poly.fit_transform(df[primary_features])

# Convert pairwise interactions to DataFrame
interaction_df_2 = pd.DataFrame(interaction_features_2, columns=poly.get_feature_names_out(primary_features))

# Generate interaction features for up to 3 columns
interaction_features_3 = {}
for comb in combinations(primary_features, 3):
    interaction_feature_name = ' '.join(comb)
    interaction_features_3[interaction_feature_name] = df[list(comb)].prod(axis=1)

interaction_df_3 = pd.DataFrame(interaction_features_3)

# Generate interaction features for up to 4 columns
interaction_features_4 = {}
for comb in combinations(primary_features, 4):
    interaction_feature_name = ' '.join(comb)
    interaction_features_4[interaction_feature_name] = df[list(comb)].prod(axis=1)

interaction_df_4 = pd.DataFrame(interaction_features_4)

# Combine all interaction DataFrames
interaction_df = pd.concat([interaction_df_2, interaction_df_3, interaction_df_4], axis=1)

# Add the primary features back to the DataFrame
interaction_df[primary_features] = df[primary_features]

# Add the workclass feature to the DataFrame
interaction_df[workclass_feature] = df[workclass_feature]

# Add the target variable
interaction_df['income'] = df['income']

# Select only the relevant interaction features (top correlations from the previous analysis)
selected_interactions = [
    'age education-num', 'education-num hours-per-week', 'age hours-per-week',
    'age capital-gain', 'education-num capital-gain', 'hours-per-week capital-gain',
    'education-num capital-loss', 'hours-per-week capital-loss', 'age capital-loss',
    # Add three and four column combinations if necessary
    'age education-num hours-per-week', 'age education-num capital-gain', 'education-num hours-per-week capital-gain',
    'age education-num capital-loss', 'education-num hours-per-week capital-loss', 'age hours-per-week capital-loss',
    'age education-num hours-per-week capital-gain', 'age education-num hours-per-week capital-loss'
]

# Final feature set
final_features = primary_features + workclass_feature + selected_interactions

# Subset the DataFrame to include only the final features and the target variable
final_df = interaction_df[final_features + ['income']]

# Display the final DataFrame
final_df.head()
```
